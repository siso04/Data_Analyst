### 4.1 Para filtrar por varias categorias creamos una lista y validamos con "isin()"

```python
# Para filtrar por multiples categorias, podemos crear una lista
# Hagamos un ejemplo, selecionando solo los paises que aparecen al principio
countries = ['United States of America', 'Austria', 'Norway', 'Taiwan']
```

```python
countries
```

```python
# Ahora aplicamos un doble filtro: primero comprabamos que los paises esten en la lista, y luego
# repetimos el filtro de los que ganan mas de 150000 dolares
paises = (df['Country'].isin(countries)) & (df['CompTotal'] >= 150000)
```

```python
df.loc[paises, ['Country', 'CompTotal']]
```

### 4.2 Filtrar con metodos de string

```python
# Primero buscamos en el esquema cual es la pregunta asociada a los lenguajes que utilizan las personas
schema_df.loc['Language', 'question']
```

```python
# Ahora buscamos en el df donde se almacenan los resultados de esa pregunta
df['LanguageHaveWorkedWith']
```

```python
# Queremos saber Quienes incluyen python entre los lenaguajes de programacion que utilizan?
# Para ello utilizaremos un filtro, aplicando el metodo de string "contains": str.contains('texto')
# Que nos indicara, si dentro de los lenguajes que utilizan las personas, se encuentra python

lenguajes = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
# Ponemos na=False, para no tener problemas con los valores nulos
```

```python
lenguajes
```

```python
# Ahora vamos aplicar el filtro a nuestro data frame 
df.loc[lenguajes, ['LanguageHaveWorkedWith', 'CompTotal']]
```

### 8.1 Convetir un objeto a float con "astype()"

```python
# Necesitamos convertir la variable "YearsCode" a flotante para poder hacer calculos con ella
df['YearsCode'] = df['YearsCode'].astype(float) 

# Aparece un error: could not convert string to float: 'More than 50 years'
```

```python
# Hay valores de texto que estan arrojando error, debemos saber cuales son para poder sustituirlo por un valor valido
df['YearsCode'].unique()
```

```python
# Tenemos dos valores de texto: 'More than 50 years' y 'Less than 1 year' que debe ser sustituidos por un numero
df['YearsCode'].replace('Less than 1 year', 0, inplace = True)
df['YearsCode'].replace('More than 50 years', 51, inplace = True)
```

```python
# Ahora veamos otra vez los valores unicos de la columna
df['YearsCode'].unique()
```

```python
# Reemplzamos los textos, pero sigue siendo un "dtype=object", tenemos que pasarlo a float
df['YearsCode'] = df['YearsCode'].astype(float) 
```

```python
# Ahora si podemos proceder a realizar calculos numericos
df['YearsCode'].mean()
```
