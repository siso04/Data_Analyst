## En esta lección, aprendermos como guardar (write) datos en diferentes formatos

```python
# Primero importamos los paquetes
import numpy as np
import pandas as pd
```

```python
# Para poder cargar un archivo .csv, utilizamos la funcion "read_csv('ruta del archivo')"
df = pd.read_csv('data/survey_results_public.csv')
df
```

```python
df.set_index('ResponseId', inplace = True)
df.head(5)
```

## 1. ¿Cómo exportar un archivo .csv? df.to_csv

```python
# Lo primero que vamos es filtrar, por un pais, para que nuestra base de datos no sea tan pesada
filt = (df['Country'] == 'United States of America')
usa_df = df.loc[filt]
```

```python
usa_df.head(5)
```

```python
# Ahora si podemos exportar nuestra tabla filtrada, como archivo .csv
usa_df.to_csv("data/usa.csv")
```

## 2. ¿Como exportar un archivo de data tabular? 

```python
# El archivo de data tabular utiliza la misma función "df.to_csv" solo que cambiamos la extension "tsv"
# y agremaos el argumento de separación
usa_df.to_csv("data/usa.tsv", sep = "\t")
```

## 3. ¿Como exportar a un archivo de Excel? df.to_excel()

```python
usa_df.to_excel("data/usa.xlsx")
```

## 4. ¿Como leer/cargar un archivo de Excel? pd.read_excel()

```python
tabla_excel = pd.read_excel("data/usa.xlsx", index_col = 'ResponseId')
```

```python
tabla_excel.head(6)
```

## 5. ¿Como exportar a un archivo a JSON? df.to_json()

```python
# La estructura que se utiliza es la misma, que para otras exportaciones
usa_df.to_json("data/tabla_json.json")
# La salida es un archivo en formato de diccionario
```

```python
# Si pasamos el argumento "orient = 'Records'" obtendremos data tabular en lugar de diccionarios
usa_df.to_json("data/tabla_json1.json", orient = 'records', lines = True)
```

## 6. ¿Como leer/carga un archivo a JSON? pd.read_json()

```python
tabla_json = pd.read_json("data/tabla_json1.json", orient = 'records', lines = True)
```

```python
tabla_json
```

## 7. Cargar tablas desde una base de datos SQL - df.to_sql

```python
from sqlalchemy import create_engine # Se importa el paquete
engine = create_engine('sqlite://', echo=False) # Se crea un engine para conectar la base de datos
```

```python
# Ahora creamos un tabla de ejmplo, que luego mandaremos a una base de SQL
df = pd.DataFrame({'name' : ['User 1', 'User 2', 'User 3']})
df
```

```python
# Ahora pasmos nuestra tabla sencilla a sql
df.to_sql('users', con=engine)
```

```python
engine.execute("SELECT * FROM users").fetchall()
```

## 8. Leer tablas que estan en una base de datos de SQL - pd.read_sql

```python
# Para este ejemplo, tenemos que exportar SQLITE3
from sqlite3 import connect
conn = connect(':memory:')
df = pd.DataFrame(data=[[0, '10/11/12'], [1, '12/11/10']],
                  columns=['int_column', 'date_column'])
```

```python
df.to_sql('test_data', conn)
```

```python
# Ahora podemos leer nuestra tabla de datos
pd.read_sql('SELECT int_column, date_column FROM test_data', conn)
```
