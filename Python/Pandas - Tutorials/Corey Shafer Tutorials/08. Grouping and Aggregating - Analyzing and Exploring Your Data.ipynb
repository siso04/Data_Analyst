## 3. Agrupaciones - df.groupby()

```python
# Vamos a realizar agrupaciones por pais, asi que veamos la columna 'Country'
df['Country'].unique()
```

```python
# Ahora veamos cuantas personas de cada pais respondieron la encuesta
df['Country'].value_counts()
```

### 3.1 Como generar una agrupacion

```python
# Utilizaremos la funcion groupby() y escogeremos una columna de nuestra base de datos
# Esta es solo la operacion
df.groupby(['Country'])
```

```python
# Para poder aplicarlo, debemos asignarlo a una variable
country_grp = df.groupby(['Country'])
```

```python
# Una vez que hemos definido el grupo podemos seleccionar elementos y aplicar metodos de grupo
country_grp.get_group('Germany').head(5)
# Todo lo que veamos estara filtrado por ese elemento de grupo 'Germany'
```

### 3.2 Generar calculos con una agrupacion

```python
# Para generar calculos, solo debemos tomar nuestra agrupacion y alguna columna, y le aplicamos un metodo
country_grp['MiscTechHaveWorkedWith'].value_counts()
# Esto nos ofrece un total de esta variable, por cada elemento del grupo (en este caso de cada pais)
```

```python
country_grp['CompTotal'].median()
```

### 3.3 Realizar seleccion con ".loc"

```python
# Las agrupaciones tienen varios indices, pero el primer indice, siempre sera la variable de agrupacion
# en este caso 'Country', si queremos seleccionar un elemento solo debemos utilizar el indice (en este caso
# el indice es el nombre del pais)
country_grp['MiscTechHaveWorkedWith'].value_counts().loc['China']
```

```python
# Podemos hacer el mismo calculo con porcentajes "normalize=True"
country_grp['MiscTechHaveWorkedWith'].value_counts(normalize=True).loc['India']
```

```python
country_grp['CompTotal'].median().loc['Germany']
```

### 3.4 Aplicar multiples funciones de agregacion - agg()

```python
# Si queremos, podemos aplicar multiples funciones de agregacion a todo el grupo o a un solo individuo
# Para ello utilizaremos la funcion "agg()" y una lista de metodos [ ]
# Ejemplo para todos los elementos del grupo
country_grp['CompTotal'].agg(['median', 'mean', 'std'])
```

```python
# Ejemplo para un solo individuo
country_grp['CompTotal'].agg(['median', 'mean', 'std']).loc['Canada']
```

### 3.5 Aplicar funciones a todo la Serie con una funcion "lambda"

```python
# En los objetos groupby no se pueden aplicar funciones como 'str' de forma directa, debe utlizarse
# el metodo apply() y una funcion lambda
country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum()).sort_values(ascending=False)
```

## 4. Ejemplo: Como determinar porcentaje de personas que utilizan Python?

```python
# Primero vamos a ver cuantas personas de cada pais respondieron a la encuentas
countries = df['Country'].value_counts()
countries
```

```python
# Ahora vamos a determinar con un metodo 'str' y 'sum()' cuantas personas usan Python en cada pais
countries_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
countries_python
```

### 4.1 Combinar Series con "pd.concat()"

```python
# Cada uno de los calculos que hicimos arriba, arroja una Serie cada uno, para poder hacer el analisis
# debemos hacer un Data Frame con esas dos Series
# IMPORTANTE!!! Utilizar 'axis='columns' para que haga el enlace por la columna, porque por defecto lo hacer por fila
df_python = pd.concat([countries, countries_python], axis='columns', sort=False)
df_python
```

```python
# Vamos a cambiar los nombres de las columnas para que tenga mas sentido el analisis
df_python.rename(columns={'Country':'Total_Encuestados', 'LanguageHaveWorkedWith':'Utiliza_Python'}, inplace=True)
df_python
```

### 4.2 Calcular el porcentaje

```python
# Crearemos una nueva columna, que aparecera al final y tendra el calculo del porcentaje
df_python['Porcentaje'] = (df_python['Utiliza_Python']/df_python['Total_Encuestados'])*100
df_python
```

```python
# Ahora podemos hacer calculos con esa nueva variable
df_python.sort_values(by='Porcentaje', ascending=False, inplace=True)
df_python
```
