# Archivos de Excel

#### El método read_excel() puede leer archivos de Excel 2007+ (.xlsx) utilizando el módulo de Python openpyxl. Los archivos de Excel 2003 (.xls) se pueden leer con xlrd. Los archivos binarios de Excel (.xlsb) se pueden leer con pyxlsb. El método de instancia to_excel() se usa para guardar un DataFrame en Excel. En general, la semántica es similar a trabajar con datos csv

#### Seleccionar columnas específicas

```python
# Suele ocurrir que los usuarios insertan columnas para realizar cálculos temporales en Excel y es posible 
# que no desee leer en esas columnas. read_excel toma una palabra clave usecols para permitirle especificar 
# un subconjunto de columnas para analizar.

# pd.read_excel("path_to_file.xls", "Sheet1", usecols="A,C:E") -> Orden en Excel

# pd.read_excel("path_to_file.xls", "Sheet1", usecols=[0, 2, 3]) -> Por indice

# pd.read_excel("path_to_file.xls", "Sheet1", usecols=["foo", "bar"]) -> Por nombre
```
