### Como definir el separador de los datos tabulares?

```python
movies = pd.read_table('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user')
movies
```

```python
# Debemos definir el separador pipeline, antes de abrir la tabla
movies = pd.read_table('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user', sep = '|')
movies
```

#### Como definir los titulos de las columnas?

```python
# Como nuestra tabla no tiene titulo de columnas, debemos definirlos previamente al momento de cargarla
# Para ello utilizaremos "header=None"
movies = pd.read_table('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user', sep = '|', 
                      header=None)
movies
```

```python
# Podemos definir los nombres las columnas, con el metodo "columns" y una lista
movies.columns = ['ID', 'Edad', 'Genero', 'Ocupacion', 'Codigo_Postal']
movies
```

#### Tenemos que agregar los nombres de las columnas

```python
# Para agregar los nombres de las columnas, creamos una lista antes de cargar la tabla, y utilizamos el argumento "names"
columnas = ['ID', 'Edad', 'Genero', 'Ocupacion', 'Codigo_Postal']
movies1 = pd.read_table('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user', sep = '|', 
                      header=None, names=columnas)
movies1
```
