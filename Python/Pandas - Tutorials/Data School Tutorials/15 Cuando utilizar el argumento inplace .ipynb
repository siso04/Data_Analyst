```python
import pandas as pd
```

```python
ufo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv')
ufo
```

```python
ufo.dtypes
```

### Para eliminar una columna con drop()

```python
ufo.drop(columns='City', inplace = True)
ufo
```

### Para cambiar los nombres de las columnas

```python
ufo.rename(columns={'Colors Reported':'Colors', 'Shape Reported':'Shape'}, inplace=True)
```

```python
ufo
```

### Para Ordenar Valores por una condicion

```python
ufo.sort_values(by='Shape', inplace=True)
ufo
```

### Para utlizar una columna como indice

```python
ufo.set_index("State", inplace=True)
ufo
```
