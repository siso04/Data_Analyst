```python
import pandas as pd
```

```python
ufo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv')
ufo
```

```python
ufo.columns
```

### Renombrar columnas con rename() y un diccionario - inplace=True

```python
# Para renombrar columnas utilizamos la funcion rename, y un diccionario con key:columna vieja, value:columna nueva
ufo.rename(columns={'Colors Reported':'Colors_Reported', 'Shape Reported':'Shape_Reported'}, inplace = True)
ufo
```

### Renombrar las columnas con una lista y el metodo .columns

```python
columnas = ['ciudad', 'colores_reportados', 'forma_reportada', 'estado', 'fecha']
ufo.columns = columnas
```

```python
ufo
```

### Renombrar columnas al momento de cargar la tabla - Utilizando una lista

```python
ufo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv', names=columnas, 
                 header=0)
ufo
```

### Utilizar un metodo de string (str.replace()) para cambiar elementos en los nombres de las columnas 

```python
ufo.columns
```

```python
ufo.columns = ufo.columns.str.replace('_', '-')
```

```python
ufo
```
