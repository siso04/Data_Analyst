```python
import pandas as pd
```

```python
ufo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv')
ufo
```

### Utilizando .loc

```python
ufo.loc[0:3, :]
```

```python
ufo.loc[0:3, ['City', 'State']]
```

```python
ufo.loc[0:3, 'City':'State']
```

```python
ufo.loc[ufo['City']== 'Oakland']
```

```python
ufo.loc[ufo['City']== 'Oakland', ['City', 'State']]
```

### Utilizando .iloc

```python
ufo.iloc[1:3, :]
```

```python
ufo.iloc[1:11, 0:3]
```
