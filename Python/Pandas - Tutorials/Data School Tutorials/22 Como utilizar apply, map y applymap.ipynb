```python
import pandas as pd
```

```python
train = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/titanic_train.csv')
train
```

### .map se utiliza para Series

```python
# Permite cambiar los valores de una Serie a un tipo diferente, lo que hace es aplicar una operacion sobre todos
# los elementos de la Serie
# Por ejemplo, aqui vamos a convertir la palabra female a 0, y la palabra male a 1, lo que hace .map es aplicar esta 
# operacion a todos los valores de esa Serie
train['Sex_num'] = train['Sex'].map({'female':0, 'male':1})
```

```python
train.loc[0:5, ['Sex', 'Sex_num']]
```

### .apply funciona para Series y Data Frame

```python
# apply aplica una funcion a cada elemento de la Serie o Data Frame
# Podemos aplicar por ejemplo una funcion de string y crear una nueva columna
train['name_lenght'] = train['Name'].apply(len)
```

```python
train.loc[0:5, ['Name', 'name_lenght']]
```
