```python
import pandas as pd
```

```python
movies = pd.read_table('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user', sep = '|', 
                      header=None)
movies
```

```python
movies.columns = ['ID', 'Edad', 'Genero', 'Ocupacion', 'Codigo_Postal']
movies
```

### Calcular cuantos duplicados hay en una columna con duplicated() and sum()

```python
movies.duplicated('Codigo_Postal').sum()
```

```python
movies.duplicated('ID').sum()
```

### Para saber la posicion o indice de los valores duplicados, podemos utilizar .loc

```python
# Utilizamos df.loc[] y en los corchetes colocamos la columna y la funcion con el duplicado
movies.loc[movies.duplicated('Codigo_Postal')]
```

### Los duplicados se eliminan con la funcion drop_duplicates()

```python
# Podemos aplicar esta funcion para todo el data frame
movies.drop_duplicates()
```

### Seleccionar una o varias columnas para evaluar duplicados con el argumento "subset="

```python
# Podemos evaluar si en esta combinacion de columnas hay valores duplicados
movies.duplicated(subset=['Edad', 'Codigo_Postal'])
```

```python
# Podemos sumarlos para saber cuantos son
movies.duplicated(subset=['Edad', 'Codigo_Postal']).sum()
```

```python
# Podemos eliminar solo los duplicados que se encuentren en esas dos columnas con "drop_duplicates()"
movies.drop_duplicates(subset=['Edad', 'Codigo_Postal'])
```
