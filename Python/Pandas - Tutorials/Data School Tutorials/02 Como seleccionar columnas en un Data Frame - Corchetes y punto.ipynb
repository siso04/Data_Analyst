### Como unir dos columnas con el simbolo de suma +

```python
# Para unir dos columnas solo tenemos que concatenarlas con el simbolo de suma +
ufo.City + ", " + ufo.State
```

```python
# Ahora podemos agregar esa columna al data frame
ufo['Location'] = ufo.City + ", " + ufo.State
ufo
```
