## Aparece una advertencia: Try using .loc[row_indexer,col_indexer] = value instead

```python
movies[movies['content_rating'] == 'NOT RATED']['content_rating'] = np.nan
```

### La solucion es aplicar .loc

```python
# La advertencia nos dice: debe seleccionar el indice de la fila y el indice de la columna, y luego asignar el valor
# es decir, debemos seleccionar una columna con una condicion, que es el indice, y luego repetir a columna, que es donde
# se aplicara esa condicion.
# A toda esa operacion le aplicamos el metodo con un simbolo de igual
movies.loc[movies['content_rating'] == 'NOT RATED', 'content_rating'] = np.nan
```
