### 3. Eliminar los missing values - dropna()

### 3.1 Todas las filas con valores nulos - df.dropna(how = 'any')

```python
# Para eliminar cualquier fila que contenga un NaN, utilizamos df.dropna(how = 'any')
ufo.dropna(how = 'any')
```

### 3.2 Las filas donde todos los valores sean nulos - df.dropna(how = 'all')

```python
# Para eliminar las filas, solo donde todos los valores sean nulos, utilizamos df.dropna(how = 'all')
ufo.dropna(how = 'all')
```

### 3.3 Utilizar dropna() solo en una o unas columnas especificas - subset=

```python
# Para hacer esto, aplicamos la funcion dropna(subset=) y dentro de "subset" especificamos una lista []
# con las columnas que queremos validar
ufo.dropna(subset = ['City', 'Shape Reported'], how = 'any')
```
