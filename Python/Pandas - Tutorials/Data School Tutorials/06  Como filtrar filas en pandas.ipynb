```python
import pandas as pd
```

```python
movies = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv')
movies
```

### Filtrar con un condicional y una variable

```python
# Para filtrar con un condicional, creamos una variable de filtro
filtro = (movies['duration'] >= 150)
```

```python
movies[filtro]
```

### Aplicar un filtro directo al data frame - df[(df[â€˜columnaâ€™] condicion)]

```python
# Para hacerlo, creamos el filtro y lo aplicamos directo al data frame df[(filtro)]
movies[(movies['duration'] >= 150)]
```

### Filtrar con .loc y dos o mas variables

```python
# Filtrar con .loc y dos o mas variables
movies.loc[filtro, ['title','star_rating','duration']].sort_values(by='star_rating', 
                                                                   ascending = False).head(20)
```

### Aplicar filtros dobles - Dos condiciones & u |

```python
filtro = (movies['duration'] >= 180) & (movies['genre'] == 'Drama')
```

```python
movies[(movies['duration'] >= 180) & (movies['genre'] == 'Drama')]
```

```python
movies[(movies['duration'] >= 180) | (movies['genre'] == 'Drama')]
```

### Filtrar multiples condiciones de OR con isin()

```python
# Creamos las condiciones del filtro, pero dentro de la funcion isin()
movies[(movies['genre'].isin(['Crime', 'Drama', 'Adventure']))]
```
