```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (10, 5)
```

Se crea un gráfico de líneas con la función **plot** y representa una serie diferente de puntos de datos y conecta cada punto en una serie con una línea.

La funcion **plot** es lo suficientemente inteligente como para darse cuenta de que lo que queremos usar es el índice de la serie como el valor de x. Lo cual es bastante útil cuando quieres hacer gráficos rápidos.

En segundo lugar, vemos que la gráfica identifica esto como dos series de datos y que los colores de los datos de la serie son diferentes, incluidos los puntos de datos y las líneas entre los puntos de datos.

```python
# Crearemos una serie de datos
linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2

# Esta vez usaré el lenguaje de formato mini para describir cómo quiero los marcadores
# y líneas a representar. La -o significa usar una línea sólida con marcadores circulares.
# Tenga en cuenta que estoy pasando los datos seguidos del formato para cada serie, por lo que
# veremos el resultado como una serie de dos datos, el lineal en la parte inferior y el cuadrático uno en la parte superior.

plt.figure()
plt.plot(linear_data, '-o', exponential_data, '-o');
```

Es bastante fácil extender esto y agregar una nueva línea arbitraria, esta vez **voy a usar --r que para trazar una línea discontinua en rojo**.

```python
plt.plot(linear_data, '-o', exponential_data, '-o')
plt.plot([22,44,55], '--r')
```

### Crear la leyenda y las etiquetas

Podemos usar las funciones de ejes regulares creando etiquetas para los ejes y para la figura como un todo. Y también podemos crear una leyenda. Pero como no etiquetamos los puntos de datos como lo hicimos con el diagrama de dispersión, necesitamos crear entradas de leyenda cuando agregamos la leyenda misma.

```python
# Crear las etiquteas
plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('Graphic title')

# Ahora creamos una leyenda con entradas de leyenda (porque no teníamos etiquetas cuando trazamos la serie de datos)
plt.legend(['Base line', 'Competition', 'Us'])
plt.plot(linear_data, '-o', exponential_data, '-o')
plt.plot([22,44,55], '--r')
```

### Rellenar el area entre las lineas con "fill_between"

Este es un buen momento para introducir la función de relleno en matplotlib. Ahora bien, esta función no es específica de los diagramas de líneas, pero puede ver que se usa comúnmente con estos diagramas, así que la presentaré en este contexto. Imaginemos que queremos resaltar la diferencia entre las curvas verde y azul. Podríamos decirle a los ejes que pinten un color entre estas series usando la función **fill_between**.

```python
# Agregamos las etiquetas y la leyenda como hicimos anteriormente
plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('A title')
plt.legend(['Baseline', 'Competition', 'Us'])
plt.plot(linear_data, '-o', exponential_data, '-o')

# Ahora necesitamos obtener el objeto de ejes actual y llamar a fill_ between. No especificamos ninguna x
# valores en nuestra llamada a plot, por lo que solo usaremos el mismo rango de puntos de datos que ya está usando.
# Luego pondremos nuestros límites inferiores y nuestros límites superiores junto con el color que queremos pintar
# y por diversión, incluiré un valor de transparencia. Y eso es todo lo que hay que hacer.
plt.gca().fill_between(range(len(linear_data)),
                      linear_data, exponential_data,
                      facecolor = 'red',
                      alpha=0.25)
```

### Trabajar con fechas

```python
# Usaré la función de organización NumPy que viste anteriormente es muy útil para generar
# algunas fechas de muestra para nosotros. Podemos pedir un arreglo de fechas a intervalos diarios solo
# proporcionando las fechas de inicio y finalización.
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype = 'datetime64[D]')

# Ahora podemos volver a trazar nuestros datos lineales y exponenciales contra este conjunto de fechas.
plt.plot(observation_dates, linear_data, '-o', observation_dates, exponential_data, '-o')
```

### Modificar y editar los ticks del grafico

Sin embargo, lo que quiero mostrarte es que puedes acceder al texto de los tics usando la función **get tick labels**. Cada una de las etiquetas de marca es un objeto de texto que en sí mismo es un artista. Esto significa que puede usar varias funciones de artista diferentes y una específica para el texto es la función de rotación establecida que cambia la rotación en función de los grados. Recorramos las etiquetas de los ejes y cambiemos eso.

```python
# Este es nuestro grafico
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype = 'datetime64[D]')
plt.plot(observation_dates, linear_data, '-o', observation_dates, exponential_data, '-o')

# Ahora creamos una variable que apunte al eje x
# Con la cual rotaremos las etiquetas del grafico en 45 grados
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
```

Ahora vamos agregar las etiquetas

```python
# Merging code from previous
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o');

x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)

# Agregamos las etiquetas
ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')
ax.set_title('Exponential vs Linear Performance')
```

###  Definir el tamano de la figura con: figsize=( , )

```python
# Al crear una nueva figura con el parámetro figsize le digo a matplotlib
# que quiero una figura de 8x6 pulgadas a 300 DPI
plt.figure(figsize=(8,6))

# Now we can bring in our plotting details from before
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o');
x = plt.gca().xaxis
# rotate the tick labels for the x axis
for item in x.get_ticklabels():
    item.set_rotation(45)
ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')

# Y finalmente, añadiré algunas expresiones matemáticas que puedes hacer
# en cualquier elemento de texto
ax.set_title("Exponential ($x^2$) vs. Linear ($x$) performance")
```
