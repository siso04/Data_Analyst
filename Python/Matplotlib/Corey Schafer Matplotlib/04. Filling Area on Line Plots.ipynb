```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-dark-palette')
```

```python
# Tabla de datos
data = pd.read_table('data.txt', sep = ',')
data.head(10)
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.figure(figsize=(8,6))
plt.plot(ages, dev_salaries, color='gray', 
         linestyle='--',
        label='All Devs')
plt.plot(ages, py_salaries, color='r', 
         linestyle='-.',
        label='Python')
plt.title('Median salary for developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.tight_layout()
plt.legend()
```

### 1. Rellenar el area a partir de una linea con plt.fill_between()

```python
plt.figure(figsize=(8,6))
plt.plot(ages, dev_salaries, color='black', 
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='r', 
         linestyle='-.',
        label='Python')
# Creamos un area de relleno
plt.fill_between(ages, py_salaries, color='r', alpha=0.3)

plt.title('Median salary for developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.tight_layout()
plt.legend()
```

### 2. Utilizar un valor para establecer la condicion del area - parametro "y2"

```python
# Vamos a utilizar como valor de referencia el salario medio de todos los programadores
# para establecer dentro del area del grafico un punto de referencia # overall_median = 57287
# El nuevo parametro se agrega como un variable con un valor
plt.figure(figsize=(8,6))
plt.plot(ages, dev_salaries, color='black', 
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='r', 
         linestyle='-.',
        label='Python')
# Creamos una variable para definir un punto de referencia
overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, color='r',  alpha=0.3)

plt.title('Median salary for developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.tight_layout()
plt.legend()
```

### 3. Definir los colores para cada categoria, por debajo o por encima del limite con "where"

```python
plt.figure(figsize=(8,6))
plt.plot(ages, dev_salaries, color='black', 
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='r', 
         linestyle='-.',
        label='Python')
# Para definir los valores de una categoria utilizamos el argumento "where" y definimos la condicion
overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries > overall_median), color='g',  alpha=0.3)

# Si queremos agregar otra condicion, debemos definir otro fill_between con la nueva condicion
plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries <= overall_median), color='r',  alpha=0.3)

plt.title('Median salary for developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.tight_layout()
plt.legend()
```

### 4.  Rellenar el area entre las dos lineas y agregar las etiquetas

```python
# Para rellenar el area entre las dos lineas, sencillamente el lugar de utilizar un valor fijo
# debemos utlizar una de las series de datos que graficamos, en el argumento "where"
# Aqui vamos a comparar dev_salaries con py_salaries

plt.figure(figsize=(8,6))
plt.plot(ages, dev_salaries, color='black', 
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='r', 
         linestyle='-.',
        label='Python')
# En lugar de comparar con valor fijo, comparamos con otra serie de datos: dev_salaries
overall_median = 57287
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries > dev_salaries), color='g',  alpha=0.3, label= "Above median")

# Si queremos agregar otra condicion, debemos definir otro fill_between con la nueva condicion
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries <= dev_salaries), color='r',  alpha=0.3, label= "Below median")

plt.title('Median salary for developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.tight_layout()
plt.legend()
```
