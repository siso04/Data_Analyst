```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# Debemos importar dos modulos adicionales para las fechas
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
```

```python
# Tabla de datos
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

### 1. Utilizar plot_dates para plotear las fechas

```python
# Por defecto aparecen puntos, debemos agregar el atributo "linestyle" para que aparezca la linea
plt.plot_date(dates, y, color='red', linestyle='solid')
plt.tight_layout()
```

### 2. Ajustar las etiquetas de la fecha con plt.gcf().autofmt_xdate()

```python
plt.plot_date(dates, y, color='red', linestyle='solid')
plt.gcf().autofmt_xdate() # Coloca las etiquetas en un angulo inclinado
plt.tight_layout()
```

### 3. Ajustar el formato de las fechas

```python
from matplotlib import dates as mpl_dates
```

```python
date_format = mpl_dates.DateFormatter('%b, %d, %Y')
```

### 4. Ajustar el formato de fecha al grafico con plt.gca().xaxis.set_major_formatter()

```python
# Para ajustar el nuevo formato de fecha al grafico, debemos ajustar el eje y colocar el nuevo formato de fecha
plt.plot_date(dates, y, color='red', linestyle='solid')
plt.gcf().autofmt_xdate() 
plt.gca().xaxis.set_major_formatter(date_format)
plt.tight_layout()
```

### 5. Ejemplo con datos reales

```python
data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/08-TimeSeries/data.csv')
data.head()
```

```python
price_date = data['Date']
price_close = data['Close']
```

```python
plt.plot_date(price_date, price_close, linestyle='solid', color='black')
plt.gcf().autofmt_xdate() 
```

### 6. Los datos de la fecha estan en formato texto, debemos pasarlos a formato fecha con Pandas

```python
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
```

```python
plt.plot_date(price_date, price_close, linestyle='solid', color='black')
plt.gcf().autofmt_xdate() 
```
