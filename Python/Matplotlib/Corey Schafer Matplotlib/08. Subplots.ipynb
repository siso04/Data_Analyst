```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# Datos
data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/10-Subplots/data.csv')
data.head()
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

### 1. Crear subplots

Para crear subplots se utiliza la siguiente convencion:

    * fig, ax = plt.subplots()

**fig**: representa a la figura general, o al marco de datos. Digamos que es como el marco del dibujo.

**ax**: representa el contenido interno, es decir, la cantidad de graficos que hay dentro del marco. Se definen por filas y columnas.. Por ejemplo, si "ax" es igual a 2, tendremos dos graficos dentro de esa figura.

### 2. Reemplazar plt por ax, para configurar los subplots

Para poder utilizar los subplots, **debemos reemplazar "plt" por "ax", para poder referenciar sobre la figura**, que ahora se refiere a un eje particular, y no a un solo grafico. En el caso de los titulos y las etiquetas, **debemos utilizar "set_title" y "set_label"**.

```python
fig, ax = plt.subplots()

# Sustituimos ax por plt
ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')
ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax.legend()

# Debemos utilizar set_
ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

# Este no se cambia
plt.tight_layout()
```

### 3. Creacion de los subplots: agregar filas y columnas a plt.subplots()

Para poder crear los subplots debemos **agregar otro eje "ax" y definir la cantidad de filas y columnas** del grafico

```python
# Los subplots estan representados por "ax1 y ax2", mientras que dentro de la funcion 
# decidimos como van a mostrarse, en este caso dos filas y una columna
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
```

### 4. Debemos **renombrar los ejes ax1 y ax2**, para asignarlos a cada grafico

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

# Asignamos ax1 y ax2 al grafico correspondiente
ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python', color='yellow')
ax2.plot(ages, js_salaries, label='JavaScript', color='red')

# Debemos hacer una leyenda para cada grafico, pero recordemos que hay elementos que coninciden
# como por ejemplo los ejes x e y del grafico
# En este caso utilizaremos un solo titulo para los dos graficos
# y el eje x de datos del primer grafico lo eliminaremos para que no se repita
# Lo ultimo que haremos es compartir el eje x, para que solo aparezca abajo de los dos graficos
# esto se hace arriba en la funcion subplots: sharex=True

ax1.legend()
ax1.set_title('Median Salary (USD) by Age', fontsize=15, fontweight = 'bold')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

# Este no se cambia
plt.tight_layout()
```

### 5. Desempaquetar los graficos: hacer dos graficos individuales

Para hacer dos graficos individuales, creamos una figura nueva y le asignamos un numero. El primer grafico seria el fig1 y el segundo el fig2

```python
# Asi creamos dos graficos, cada fig tiene un numero y su ax correspondiente
fig1, ax1= plt.subplots()
fig2, ax2= plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax2.plot(ages, py_salaries, label='Python', color='yellow')
ax2.plot(ages, js_salaries, label='JavaScript', color='red')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age', fontsize=15, fontweight = 'bold')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

# Este no se cambia
plt.tight_layout()
```
