```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
# Datos

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

### 1. Crear un scatter plot basico

```python
plt.scatter(x, y, color='r')
plt.title('Trending Youtube Videos')
plt.xlabel('View count')
plt.ylabel('Total likes')
```

### 2. El argumento "s" (size) permite cambiar el tamano de los puntos, con "marker" el tipo de marcador, con "edgecolor" el color de borde del circulo, y con "linewidth", el ancho de la linea del circulo

```python
plt.scatter(x, y, s=100, marker= "P", color='r', edgecolor='white', linewidth=2, alpha=0.7)
plt.title('Trending Youtube Videos')
plt.xlabel('View count')
plt.ylabel('Total likes')
plt.tight_layout()

plt.show()
```

### 3. Definir catergorias a traves de los colores  con el argumento "c"

Podemos definir categorias a traves de los colores, asignandoles puntos y valores a cada uno de ellos. En este caso, crearemos una lista de colores que vamos agregarla al grafico, para que cada uno de los puntos pertenezca a una puntuacion especifica. **Debemos utilizar el argumento "c" no la palabra color**.

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, marker= "P", c=colors, edgecolor='white', linewidth=2, alpha=0.7)
plt.title('Trending Youtube Videos')
plt.xlabel('View count')
plt.ylabel('Total likes')
plt.tight_layout()

plt.show()
```

### 4. Colorear los puntos de las categorias "c" con colormap "cmap"

Los valores o condiciones de los scatterplots no pueden colorearse de forma individual, debe utilizarse colormap, a traves del argumento "cmap".

Estos son los valores aceptados: plot_color_gradients('Sequential',
                     ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
                      'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
                      'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn'])
                      
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, marker= "P", c=colors, cmap='PuRd', edgecolor='white')
plt.title('Trending Youtube Videos')
plt.xlabel('View count')
plt.ylabel('Total likes')
plt.tight_layout()

plt.show()
```

### 5. Agregar una barra de color, para indicar la leyenda e intensidad de los colores con el argumento plt.colorbar()

```python
plt.scatter(x, y, s=100, marker= "P", c=colors, cmap='PuRd', edgecolor='white')
plt.title('Trending Youtube Videos')
plt.xlabel('View count')
plt.ylabel('Total likes')

# Agregamos la barra de colores
cbar = plt.colorbar()
# Definimos una etiqueta para esa barra
cbar.set_label('Satisfaction')

plt.tight_layout()
plt.show()
```

### 6. Cambiar el tamano de los puntos, con una lista y el argumento size

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
          538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, cmap='PuRd', edgecolor='white')
plt.title('Trending Youtube Videos')
plt.xlabel('View count')
plt.ylabel('Total likes')

# Agregamos la barra de colores
cbar = plt.colorbar()
# Definimos una etiqueta para esa barra
cbar.set_label('Satisfaction')

plt.tight_layout()
plt.show()
```

### 7. Ejemplo con datos reales

```python
data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv')
data.head()
```

```python
view_count = data['view_count']
likes= data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count, likes, c=ratio, edgecolor='white', cmap='jet')
plt.title('Trending Youtube Videos')
plt.xlabel('View count')
plt.ylabel('Total likes')

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.xscale('log')
plt.yscale('log')

plt.tight_layout()
plt.show()
```
