```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Aquí hay cinco lenguajes diferentes, Python, SQL, Java, C++ y JavaScript.
# Encontraremos sus posiciones como un rango usando NumPy como una función de rango.
# Y aquí está la popularidad valores de Stack Overflow.
languages = ['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# Crearemos un gráfico de barras basado en el rango y popularidad, luego agrega marcas x e y y
# estableceremos un título.
plt.figure(figsize=(10,8))
plt.bar(pos, popularity, align='center', color = "red", alpha=0.65)

# Se utiliza "xticks" para ordenar el orden de las barras
plt.xticks(pos, languages)
plt.ylabel('% Popularity')
plt.title('Top 5 Languages for Math and Data \nby % Popularity on Stack Overflow')
```

### Eliminar el marco del grafico con "set_visible" y un bucle for

Bien, aquí está el primer desafío: nuestra trama tiene este marco a su alrededor, pero no es realmente necesario y parece un poco pesado. Quitemos esa tinta. Esto es un poco más complicado, pero podemos obtener el eje actual, luego iterar a través de toda la columna vertebral, configurando su visibilidad en falso. Eso ya hará que el gráfico se vea mucho más ligero.

```python
# Copiamos el codigo inicial del grafico 
import matplotlib.pyplot as plt
import numpy as np

languages = ['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

plt.figure(figsize=(10, 8));
plt.bar(pos, popularity, align='center');
plt.xticks(pos, languages);
plt.ylabel('% Popularity');
plt.title('Top 5 Languages for Math & Data \nby % Popularity on Stack Overflow');

# Eliminar el marco del grafico
for spine in plt.gca().spines.values():
    spine.set_visible(False)
```

### Cambiar los colores del grafico

Ahora, el azul está bien, pero en realidad no nos ayuda a diferenciar entre las barras en absoluto. ¿Qué tal si suavizamos todos los negros duros a gris y luego cambiamos los colores de la barra a gris también? Además, mantengamos la barra de Python del mismo color azul que tenía originalmente para que se destaque.

```python
plt.figure(figsize=(10, 8));
languages = ['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]


# Hay varias maneras diferentes en las que podemos hacer esto.
# La forma que elegí fue agregar un parámetro alfa a todo,
# que agrega un poco de transparencia y suaviza un poco los colores.
# También configuré las barras en un color gris neutro y luego elegí
# un bonito azul del sitio web de Python para acentuar la primera barra.

# cambiar el color de las barras para que tengan un color tenue
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='#aca9b0')

# cambiar una barra, la barra de python, a un color contrastante
# seleccionamos la barra como un elemento de la lista que creamos arriba
bars[0].set_color('#fc3b2d')

# suavizar todas las etiquetas volviéndose grises
plt.xticks(pos, languages, alpha=0.8)
# definir el titulo
plt.title('Top 5 Languages for Math & Data \nby % Popularity on Stack Overflow')

# Eliminar el marco del grafico
for spine in plt.gca().spines.values():
    spine.set_visible(False)
```

### Eliminar las etiquetas por defecto, y colocar etiquetas individuales para cada grafico

Ahora arreglemos el eje y eliminando las etiquetas y simplemente etiquetando directamente las barras individuales. Realmente no necesitamos la etiqueta del eje y, ya que el título nos dice todo lo que necesitamos saber sobre las unidades en este gráfico.

```python
# Retomamos el grafico que realizamos mas arriba
plt.figure(figsize=(10, 8));
languages = ['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')
bars[0].set_color('#fc3b2d')

plt.xticks(pos, languages, alpha=0.8)

# Podemos eliminar la etiqueta y simplemente configurándola en una lista vacía
plt.ylabel([])

# Agregamos el titulo y quitamos el marco
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.9)
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Quitar la etiqueta es fácil, pero cambiar las barras es un poco doloroso.
# Para esto queremos iterar sobre cada de las barras y tomar su altura.
# Entonces queremos crear un nuevo texto objeto con la información de los datos.

# Desafortunadamente, esto significa hacer un poco de jugar con el relleno.
# Aquí configuraré la ubicación x en la barra
# x más el ancho dividido por dos y la ubicación y para estar la altura de la barra menos cinco.

# Puede parecer extraño obtener el medio de la barra en la dimensión x, pero
# eso es porque estoy configurando la etiqueta para centrarse en sí mismo, horizontalmente

for bar in bars:
    height = bar.get_height()
    plt.gca().text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 5, str(int(height)) + '%',
                   ha='center', color='w', fontsize=14)
```
