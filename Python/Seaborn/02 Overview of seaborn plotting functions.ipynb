```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
# Histograma
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
plt.grid(False);
```

```python
# kernel density estimation
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

### 1. Figure-level vs. axes-level functions


Por ejemplo, **displot() es la función de nivel de figura para el módulo de distribuciones. Su comportamiento predeterminado es dibujar un histograma**, usando el mismo código que histplot() detrás de escena:

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

Para dibujar un diagrama de densidad del kernel en su lugar, **usando el mismo código que kdeplot(), selecciónelo usando el parámetro kind**

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack", kind="kde")
                                                                                 #############
```

La característica más útil que ofrecen las funciones de nivel de figura es que pueden **crear fácilmente figuras con múltiples subtramas**. Por ejemplo, en lugar de apilar las tres distribuciones para cada especie de pingüinos en los mismos ejes, podemos **"facetarlas" trazando cada distribución en las columnas de la figura**

```python
# Agregamos el argumento "col" que divide el grafico por tres categorias de especies
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species")
                                                                 #############
```

### 2. Axes-level functions make self-contained plots

```python
f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", ax=axs[0])
sns.histplot(data=penguins, x="species", hue="species", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
```

```python
# Grafico con linea personalizada
tips = sns.load_dataset("tips")
g = sns.relplot(data=tips, x="total_bill", y="tip")
g.ax.axline(xy1=(10, 2), slope=.2, color="r", dashes=(5, 2))
```

### 3. Personalización de gráficos a partir de una función de nivel de figura

Las funciones de nivel de figura devuelven una instancia de FacetGrid, que tiene algunos métodos para personalizar los atributos de la trama de una manera que es "inteligente" con respecto a la organización de la subtrama. **Por ejemplo, puede cambiar las etiquetas en los ejes externos usando una sola línea de código: set_axis_labels**

```python
g = sns.relplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", col="sex")
g.set_axis_labels("Flipper length (mm)", "Bill length (mm)")
                            #x               #y
```
