En el espíritu de Tukey, las gráficas de regresión en Seaborn están destinadas principalmente a agregar una guía visual que ayude a enfatizar patrones en un conjunto de datos durante los análisis de datos exploratorios.

## 1. regplot y lmplot

**Funciones para dibujar modelos de regresión lineal**:
    
    * Las dos funciones que se pueden usar para visualizar un ajuste lineal son **regplot() y lmplot()**
    
En la invocación más simple, ambas funciones dibujan un diagrama de dispersión de dos variables, x e y, y **luego ajustan el modelo de regresión y ~ x** y trazan la línea de regresión resultante y un intervalo de confianza del 95% para esa regresión

Estas funciones dibujan gráficos similares, pero **regplot() es una función de nivel de ejes e lmplot() es una función de nivel de figura**. Además, **regplot() acepta las variables x e y en una variedad de formatos que incluyen matrices numpy simples, objetos pandas.Series o como referencias a variables en un objeto pandas.DataFrame** pasado a datos. Por el contrario, **lmplot() tiene datos como parámetro obligatorio y las variables x e y deben especificarse como cadenas**. Finalmente, solo lmplot() tiene hue como parámetro. Sin embargo, la funcionalidad principal es similar, por lo que **este tutorial se centrará en lmplot()**

### 1.1 Funcion regplot

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.regplot(data=tips, x='total_bill', y='tip')
```

### 1.2 Funcion lmplot()

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```
