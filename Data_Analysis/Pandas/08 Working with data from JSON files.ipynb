```python
import pandas as pd
pd.set_option('display.max_columns', 15)

# Import requests library
import requests
```

```python
# Import data from a website
# Create a variable and apply requests.get. At the end call the json() method
covid_data = requests.get('https://api.covid19api.com/summary').json()
```

```python
# The data is a JSON file with list and dictionaries
covid_data
```

```python
# We can see the keys of the dict
covid_data.keys()
```

```python
# We can select the values of one key.
# Let's take Countries for example
covid_data.get('Countries')
```

```python
# To manage the data we will create a data frama from the JSON file
# We have to select a key from the original data
covid_df = pd.DataFrame.from_dict(covid_data['Countries'])
covid_df
```

```python
covid_df.columns
```

```python
covid_df.drop(columns=["ID", 'Premium', 'Date'], inplace=True)
covid_df
```

```python
# Plot total confirmed cases by country (Top Ten)
covid_df.set_index("Country")['TotalConfirmed'].sort_values().head(10).plot.barh();
```

```python
# Plot total deaths by country (Top Ten)
covid_df.set_index("Country")['TotalDeaths'].sort_values().head(10).plot.barh(color="red");
```

```python
# Relation between confirmed cases and deaths
covid_df.plot.scatter(x='TotalConfirmed', y='TotalDeaths');
```
