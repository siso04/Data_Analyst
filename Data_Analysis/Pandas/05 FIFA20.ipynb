### 1.1 Selecciones por indice .loc y .iloc

```python
# La selecciones por indice, toman como prioridad la fila, no la columna, por esa razon, la seleccion de las columnas
# puede ser un poco confusa
# loc y iloc utilizan dos argumentos [filas, columnas], para poder seleccionar columnas, debemos colocar dos puntos ":"
# en el apartado de las filas
```

### 1.2 El slicing con .loc tiene algunas variaciones

```python
# 1. Podemos seleccionar desde una columna particular y de alli hasta el final (el final se marca con dos puntos ":")
fifa22.loc[:, 'dob':]
```

```python
# 2. Podemos hacer lo contrario, desde una marca con dos puntos ":" hasta el inicio hacia la izquierda
fifa22.loc[:, :'dob']
```

```python
# 3. Podemos seleccionar columnas de forma alternada
fifa22.loc[:, ::2] # Seleccionamos una columna y dejamos una de por medio (seleccionamos 1, 3, 5 y asi sucesivamente)
```

```python
# 4. Por ultimo podemos hacer un slicing doble, filtrando las filas con numeros y las columnas por su nombre
# en este caso, la seleccion no lleva corchetes
fifa22.loc[0:5, 'short_name':'value_eur']
```

### 1.3 Seleccion y slicing con .iloc (indice numericos de filas y columnas)

```python
# 1. Seleccionar varias columnas por su indice numerico
fifa22.iloc[:, [1,3,5]]
```

```python
# 2. Slicing de un grupo de columnas por su indice (el slicing de .iloc es excluyente, no incluye el ultimo elemento)
fifa22.iloc[:, 3:7]
```

```python
# 3. Tambien podemos seleccionar desde un punto (numero hasta el final)
fifa22.iloc[:, :8]
```

```python
# 4. Y tambien desde un punto hasta el principio (el principio se marca con un numero de indice de columna)
fifa22.iloc[:, 3:]
```

```python
# 5. Por ultimo, tambien podemos seleccionar filas y columnas por su indice [filas, columnas]. Indice excluyentes
fifa22.iloc[0:3, 0:5]
```

## 3. Filtrar filas y columnas utilizando condicionales 

### 3.1 Filtros simples: seleccion de una columna y aplicar un condicional (el resultado es un booleano)

```python
# Para filtrar solo debemos selecionar un columna df['columna'] y luego especificar la condicion: df['columna'] >= x. 
# El resultado del filtro es un valor booleano
# Ejemplo:
fifa22['nationality_name'] == 'Portugal'
```

### 3.2 Crear un filtro, para que el resultado sea una tabla y no un valor booleano

```python
# Para que no aparezca el booleano, debemos crear una variable que nos ayude a filtrar
# Se recomienda colocar el filtro en parentesis
portugal = fifa22['nationality_name'] == 'Portugal'
```

```python
# Ahora aplicamos el filtro recien creado a nuestra tabla, para que nos muestre los jugadores de Portugal
fifa22[portugal]
```

```python
# Otra forma de hacer filtros es seleccionar el data frame y dentro la consulta del filtro
fifa22[fifa22['nationality_name'] == "Portugal"].head(5)
```

```python
# Podemos aplicar filtros de cualquier tipo utilizando condicionales
# Por ejemplo, veamos todos los jugadores que tienen "overall" mayor o igual a 90
best = fifa22['overall'] >= 90
fifa22[best]
```

### 3.3 Filtrar utilizando .loc (con una variable de filtro)

```python
# Haciendo esto, obtenemos exactamente el mismo resultado de arriba
fifa22.loc[portugal]
```

### 3.4 Filtros dobles (fila y columna) con .loc

```python
# Utilizar ".loc", permite agregar al filtro una o varias columnas
# fifa22.loc[portugal, 'long_name']
fifa22.loc[portugal, ['short_name', 'overall']]
```

### 3.5 Filtrar con dos condiciones, utilizando los operadores AND (&) y OR (|) 

```python
# Cuando necesitamos filtrar por dos condiciones, podemos utilizar and y or, en pandas se utilizan & y |
# Los filtros se colocan entre parentesis
# Ejemplo con AND: filtro = (df ['last'] == 'Doe') & (df ['first] == 'John')
# Ejemplo con OR:  filtro = (df ['last'] == 'Schafer') | (df ['first] == 'John')
```

```python
# Vamos a crear un filtro doble, primero con & (donde deben cumplirse las dos condiciones)
portbest = (fifa22['nationality_name'] == 'Portugal') & (fifa22['overall'] >= 85)
fifa22[portbest]
```

```python
# Ahora vamos a crear otro con | (donde solo debe cumplirse una de las condiciones)
sueldazos = (fifa22['overall'] >= 85) | (fifa22['wage_eur'] > 200000.0)
fifa22[sueldazos].sort_values(by = 'wage_eur', ascending = False)
```

### 3.6 Negar un filtro con "~" (tilde o virgulilla)

```python
# Para negar un filtro, solo debemos colocar una virgulilla, delante de la variable de filtrado: ~filtro
# Sintaxis: df.loc[~filtro1]
# Por ejemplo, si negamos el filtro de los mejores: best = fifa22['overall'] >= 90. Vamos a tener todos los jugadores
# con un overall menor a 90
fifa22.loc[~best].head(10)
```

### 3.7 Filtrar por varias categorias/valores creamos una variable lista y validamos con "isin()"

```python
# Para filtrar por multiples categorias, podemos crear una lista
# Hagamos un ejemplo, selecionando los paises candidatos al mundial
paises = ['Argentina', 'Brazil', 'France', 'Germany', 'England']
paises
```

```python
# Ahora aplicamos un doble filtro: primero comprabamos que los paises esten en la lista, y luego
# repetimos el filtro de los que tienen mayor overall
jugadores = (fifa22['nationality_name'].isin(paises)) & (fifa22['overall'] >= 85)
```

```python
fifa22.loc[jugadores, ['short_name', 'nationality_name', 'overall', 'value_eur']].head(15)
#df.loc[paises, ['Country', 'CompTotal']]
```

### 3.8  Filtrar por metodos de string - str.contains()

```python
# Queremos saber cuales equipos incluyen en su nombre la palabra "Real"
# Para ello utilizaremos un filtro, aplicando el metodo de string "contains": str.contains('texto')
# Que nos indicara, dentro de toda la tabla, cuales paises utilizan esa palabra en el nombre del equipo

los_reales = fifa22['club_name'].str.contains('Real', na=False)
# Ponemos na=False, para no tener problemas con los valores nulos
```

```python
# Estos son los equipos que tienen la palabra "Real" en su nombre
fifa22.loc[los_reales, 'club_name'].unique()
```

## 4.3 Metodos para modificar filas de forma multiple (son mas utiles que los metodos de string)

### 4.3.1 apply (funciona en Series/columnas y Data Frame)
#### Se utiliza para aplicar una funcion a todos los valores (filas), puede utilizarse en Series y Data Frame, y tambien puede aplicarse utilizando funciones "lambda"

#### 4.3.1.1 Aplicar apply() a una Serie/Columna

```python
# apply() en Series
# Podemos utilizar apply() por ejemplo, para contar la cantidad de letras que tiene el nombre de los paises (metodo "len")
# df[columna].apply(metodo)
filtrada['nationality'].apply(len)
```

#### 4.3.1.2 Aplicar apply() a una Serie con una funcion (def)

```python
# Podemos escribir una funcion y luego pasarla con apply() a toda la columna
# Crearemos una funcion para poner el "short_name" en mayuscula
def paises (short_name):
    return short_name.upper()
```

```python
# Ahora podemos aplicar la funcion recien creada, a una columna de nuestra tabla de datos. Esto no modifica el Data Frame
filtrada['short_name'].apply(paises)
```

#### 4.3.1.3 Aplicar apply() a un Data Frame - Sobrescribir la columna

```python
# Para que la funcion se aplique al Data Frame, debemos sobrescribir la columna y aplicar la funcion "def"
filtrada['short_name'] = filtrada['short_name'].apply(paises)
filtrada
```

#### 4.3.1.4 Aplicar apply() a un Data Frame, utilizando funciones "lambda"

```python
# Podemos convertir los nombres de los paises a "Capitalize" con una funcion lambda
filtrada['nationality'] = filtrada['nationality'].apply(lambda x: x.capitalize())
```

```python
filtrada
```

```python
# Por ejemplo, podemos aplicar la funcion "len" a todo el Data Frame y saber la longitud de todas las palabras
filtrada[['nationality','short_name','long_name']].applymap(len)
```

```python
# Tambien podemos, aplicar un metodo de str, para poner todos los textos por ejemplo, en mayusculas
filtrada[['nationality','short_name','long_name']].applymap(str.upper)
```

### 11.5 Lag Plot - Gráfico para series de tiempo

```python
# Los diagramas de retraso (lag) se utilizan para verificar si un conjunto de datos o una serie de tiempo es aleatorio. 
# Los datos aleatorios no deben exhibir ninguna estructura en el diagrama de retraso. 
# La estructura no aleatoria implica que los datos subyacentes no son aleatorios.
from pandas.plotting import lag_plot
```

```python
spacing = np.linspace(-99 * np.pi, 99 * np.pi, num=1000)
data = pd.Series(0.1 * np.random.rand(1000) + 0.9 * np.sin(spacing))
```

```python
lag_plot(data);
```

### 11.6 Autocorrelation plot - Gráfico de autocorrelación

```python
# Los gráficos de autocorrelación se utilizan a menudo para comprobar la aleatoriedad en series temporales. 
# Esto se hace calculando autocorrelaciones para valores de datos en diferentes retrasos de tiempo. 
# Si la serie de tiempo es aleatoria, tales autocorrelaciones deberían estar cerca de cero para todas 
# y cada una de las separaciones de retardo de tiempo. Si la serie temporal no es aleatoria, 
# una o más de las autocorrelaciones serán significativamente distintas de cero.
# Las líneas horizontales que se muestran en el gráfico corresponden a bandas de confianza del 95 % y 99 %. 
# La línea discontinua es una banda de confianza del 99%.
from pandas.plotting import autocorrelation_plot
```

```python
spacing = np.linspace(-9 * np.pi, 9 * np.pi, num=1000)
data = pd.Series(0.7 * np.random.rand(1000) + 0.3 * np.sin(spacing))
```

```python
autocorrelation_plot(data);
```

### 11.8 RadViz - Vizualizar datos multivariables

```python
# RadViz es una forma de visualizar datos de múltiples variables. Se basa en un algoritmo simple de minimización 
# de la tensión del resorte. Básicamente configuras un montón de puntos en un plano. 
# En nuestro caso, están igualmente espaciados en un círculo unitario. Cada punto representa un único atributo. 
# Luego simula que cada muestra en el conjunto de datos está unida a cada uno de estos puntos por un resorte, 
# cuya rigidez es proporcional al valor numérico de ese atributo (están normalizados al intervalo unitario).
from pandas.plotting import radviz
```

```python
data = pd.read_csv("fifa/iris.csv")
data
radviz(data, "species");
```
